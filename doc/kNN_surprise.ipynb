```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from surprise import Reader, Dataset
from surprise.model_selection import train_test_split
from surprise import BaselineOnly, accuracy
from surprise.accuracy import rmse
from surprise import KNNBasic
```

```python
filepath = 'C:/Users/samir/Documents/code/r/ads/proj4/'
```

```python
full = pd.read_csv(filepath + 'data/ml-latest-small/ratings.csv')
```

```python
full.head()
```

```python
full.rating.unique()
```

```python
full_data = full[['userId', 'movieId', 'rating']]
```

```python
full_data.head()
```

```python
reader = Reader()
data = Dataset.load_from_df(full_data, reader)
```

```python
train, test = train_test_split(data, test_size = 0.2)
```

```python
print('Using ALS')

# define cv.function(n_epochs = 5, reg_u = 12, reg_i = 5) 
bsl_options = {'method': 'als'
               'n_epochs': 5,
               'reg_u': 12,
               'reg_i': 5}

sim_options = {'name': 'cosine',
              'user_based': False}

algo = KNNBasic(bsl_options = bsl_options, sim_options = sim_options)
```

```python
algo.fit(train)
```

```python
predictions = algo.test(test)
```

```python
predictions
```

```python
rmse(predictions, verbose= True)
```
